```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
```

# ***COLOQUE O ARQUIVO "report.csv" NA PASTA DO COLLAB***



> ### Lendo base de dados



```python
comC_df = pd.read_csv('report.csv')
```

```python
comC_df.head()
```



> ### Adicionando colunas da função de comparação 



```python
comC_df ['fleetTrucksComparison'] = (((comC_df ['MX'] - comC_df ['M']) / comC_df['M']) * 100 )
comC_df ['costComparison'] = (((comC_df ['CX'] - comC_df ['C']) / comC_df['C']) * 100 )
```

```python
display(comC_df)
```

> ### Dados referentes a frota de caminhões

```python
comC_df_trucks = comC_df [['name','M','MX']]
```

```python
comC_df_trucks.head() 
```

```python
comC_df_trucks = comC_df_trucks.melt(id_vars='name')
```

```python
display(comC_df_trucks)
```

```python
fig = px.bar(comC_df_trucks, 
             x = 'name', y = 'value',
             color = 'variable', 
             barmode = 'group',
             title = 'Comparação entre o ótimo(M) e o resultado obtido (MX)'
             )

fig.show()
```



> ### Dados referentes ao custo de rota



```python
comC_df_cost = comC_df [['name','C','CX']]
```

```python
comC_df_cost.head()
```

```python
comC_df_cost = comC_df_cost.melt(id_vars='name')
```

```python
display(comC_df_cost)
```

```python
fig = px.bar(comC_df_cost, 
             x = 'name', y = 'value', 
             color = 'variable', 
             barmode = 'group',
             title = 'Comparação entre o ótimo(C) e o resultado obtido (CX)'
             )
fig.show()
```


> ### Dados referentes a função de comparação
download.png


```python
comC_df_fleet = comC_df [['name','fleetTrucksComparison']]
comC_df_costComparison = comC_df [['name','costComparison']]
```

```python
display(comC_df_fleet)
```

```python
fig = make_subplots(rows=1, cols=2, subplot_titles = ["Frota de caminhões","Custo de rota"])

fig.add_trace(
    go.Bar(x=comC_df_fleet['name'], y=comC_df_fleet['fleetTrucksComparison']), 
    row=1, col=1)

fig.add_trace(
    go.Bar(x = comC_df_costComparison['name'], y = comC_df_costComparison ['costComparison']), 
    row=1, col=2)

fig.update_xaxes(title_text = 'Instâncias',row = 1, col = 1)
fig.update_xaxes(title_text = 'Instâncias',row = 1, col = 2)
fig.update_yaxes(title_text = 'Resultado')

fig.update_layout(height=700, width=1400, title_text= "Funções de comparação")
fig.show()
```
